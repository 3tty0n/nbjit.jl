```julia
include("../src/ijulia_integration.jl")
using .IJuliaIntegration: @jit
using Printf
```

```julia
println("=== Initial cell ===")
@jit begin
    x = 100
    @hole delta = 100
    result = x + delta
    result
end
```

```julia
println("=== Main body is reused, hole cell is compiled ===")
@jit begin
    x = 100
    @hole delta = 42 # Changed
    result = x + delta
    result
end
```

```julia
println("=== Full compilation ===")
@jit begin
    x = 100
    @hole delta = 42
    result = x + delta * 2 # Changed
    result
end
```
